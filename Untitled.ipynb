```python
# Image Augmentation
def augment_images(img):
    image = Image.fromarray(np.uint8(img))
    image = ImageEnhance.Brightness(image).enhance(random.uniform(0.8, 1.2))
    image = ImageEnhance.Contrast(image).enhance(random.uniform(0.8,1.2))
    image = np.array(image)/255
    return image

```

```python
# load Image
def open_images(paths):
    images = []
    for path in paths:
        image = load_img(path, target_size=(Image_Size, Image_Size))
        image = augment_images(image)
        images.append(image)
    return np.array(images)
```

```python
def encode_labels(labels):
    unique_labels = os.listdir(train_dir)
    encoded = [unique_labels.index(label) for label in labels]
    return np.array(encoded)
```

```python
def data_gen(paths, labels, batch_size=1, epochs=1):
    while(True):
        indices = np.arange(len(paths))
        np.random.shuffle(indices)
        shuffled_paths = [paths[i] for i in indices]
        shuffled_labels = [labels[i] for i in indices]
        for i in range(0, len(paths), batch_size):
            batch_paths = shuffled_paths[i:i + batch_size]
            batch_images = open_images(batch_paths)  # Open and augment images
            batch_labels = shuffled_labels[i:i + batch_size]
            batch_labels = encode_labels(batch_labels)  # Encode labels
            yield batch_images, batch_labels  # Yield the batch
```

```python
Image_Size = 128
```

```python
base_model = VGG16(input_shape=(Image_Size, Image_Size, 3), include_top=False, weights='imagenet')
for layer in base_model.layers:
    layer.trainable = False # freeze all layers
#unfreeze following layers
base_model.layers[-2].trainable = True
base_model.layers[-3].trainable = True
base_model.layers[-4].trainable = True

model = Sequential()
model.add(Input(shape = (Image_Size, Image_Size, 3)))
model.add(base_model)
model.add(Flatten())
model.add(Dropout(0.3))
model.add(Dense(128,activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(num_class,activation = 'softmax'))
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy',metrics=['sparse_categorical_accuracy'])

batch_size = 20
steps = int(len(train_labels)/batch_size)
epochs = 5

history = model.fit(
    data_gen(train_paths, train_labels, batch_size = batch_size, epochs=epochs),
    epochs=epochs,
    steps_per_epoch=steps
)
```

```python
def detect_and_display(img_path, model):
    try:
        img = load_img(img_path,  target_size=(Image_Size,Image_Size))
        img_array = np.array(img)/255
        img_array = np.expand_dims(img_array, axis=0) 
    
        prediction = model.predict(img_array)
        predicted_class_index = np.argmax(prediction,axis = 1)[0]
        predicted_class_name = class_labels[predicted_class_index]
        confidence_score = np.max(prediction)
    
        if predicted_class_name == 'notumor':
            result = "No Tumor"
        else:
            result = f"Tumor: {predicted_class_name}"
    
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"{result} Detected (Confidence: {confidence_score * 100:.2f})")
        plt.show()
    except Exception as e:
        print("Error processing the Image", str(e))
```

```python
import tensorflow as tf
import cv2
```

```python
def grad_cam(img_array, model, last_conv_layer_name="block5_conv3"):
    # 1. Get the VGG16 base model from the Sequential model
    base_model = model.layers[0]
    last_conv_layer = base_model.get_layer(last_conv_layer_name)
    
    # 2. Create the backbone model: maps the image to the last conv layer
    backbone = tf.keras.Model(base_model.inputs, last_conv_layer.output)
    
    # 3. Create the classifier model: maps the last conv layer to final predictions
    classifier_input = tf.keras.Input(shape=last_conv_layer.output.shape[1:])
    x = classifier_input
    
    # A. Add remaining layers from the base_model (e.g., block5_pool)
    layer_index = base_model.layers.index(last_conv_layer)
    for layer in base_model.layers[layer_index + 1:]:
        x = layer(x)
        
    # B. Add remaining layers from the outer Sequential model (Flatten, Dropout, Dense)
    for layer in model.layers[1:]:
        x = layer(x)
        
    classifier = tf.keras.Model(classifier_input, x)
    
    # 4. Compute gradients
    with tf.GradientTape() as tape:
        # Pass the image through the backbone and watch the conv outputs
        conv_outputs = backbone(img_array)
        tape.watch(conv_outputs)
        
        # Pass the watched conv outputs to the classifier to get the final predictions
        predictions = classifier(conv_outputs)
        
        # Get the highest predicted class
        class_idx = tf.argmax(predictions[0])
        loss = predictions[:, class_idx]
        
    # 5. Get gradients of the loss with respect to the conv layer outputs
    grads = tape.gradient(loss, conv_outputs)
    
    # 6. Global average pool the gradients
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # 7. Multiply each channel by its pooled gradient importance
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    
    # 8. Apply ReLU and normalize to [0, 1]
    heatmap = heatmap.numpy()
    heatmap = np.maximum(heatmap, 0)
    heatmap /= np.max(heatmap) + 1e-8  # avoid divide by zero
    
    return heatmap
```

```python
def predict_with_gradcam(img_path):

    # load image with pillow
    img = load_img(img_path, target_size=(Image_Size,Image_Size))
    img_array = np.array(img)/255.0
    img_array = np.expand_dims(img_array, axis=0)
    img_array = tf.convert_to_tensor(img_array, dtype = tf.float32)
    
    preds = model.predict(img_array)
    pred_class = np.argmax(preds,axis = 1)[0]
    confidence_score = np.max(preds)
    class_name = class_labels[pred_class]

    heatmap = grad_cam(img_array, model, "block5_conv3")
    if class_name == 'notumor':
            result = "No Tumor"
    else:
            result = f"Tumor: {class_name}"

    # original image with cv2 for overlay
    img_original = cv2.imread(img_path)
    img_original = cv2.resize(img_original,(Image_Size,Image_Size))
    img_original = img_original.astype(np.uint8)

    heatmap = cv2.resize(heatmap, (Image_Size,Image_Size))
    heatmap = np.uint8(255 * heatmap)

    img_original = img_original.astype(np.uint8)
    heatmap = heatmap.astype(np.uint8)

    
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    superimposed_img = heatmap * 0.4 + img_original

    plt.imshow(cv2.cvtColor(superimposed_img.astype("uint8"), cv2.COLOR_BGR2RGB))
    plt.title(f"{result} Detected (Confidence: {confidence_score * 100:.2f})")
    plt.axis("off")
    plt.legend()
    plt.show()
```

```python
predict_with_gradcam('MRI/Testing/meningioma/Te-aug-me_12.jpg')
```
